# Anomaly Detection in Time Series
---

> __Time series are observations that have been recorded in an orderly fashion and which are correlated in time.__

while analyzing time series data, many data analysts have to make sure of the outliers, much as we do in static data. 
We may know how much pain outliers cause for us. These outliers are called "anomalies" in time series jargon. 

## What are anomalities/outliers and types of anomalies in time-series data?

From a traditional point of view, an outlier/anomaly is : 

__"An observation which deviates so much from other obsercations as to arouse suspicions that it was generated by a different mechanism"__

Therefore, we must think of outliers as observations that don't follow the expected behavior.
